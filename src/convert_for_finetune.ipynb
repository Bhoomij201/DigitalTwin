```python
import json

# load pairs
with open("data/processed/pairs.json", "r", encoding="utf-8") as f:
    pairs = json.load(f)

output = []

for p in pairs:
    context = p["context"].strip()
    reply = p["reply"].strip()

    # skip bad data
    if not context or not reply:
        continue

    if len(reply.split()) < 2:
        continue

    # remove speaker labels if present
    if ":" in context:
        context = context.split(":", 1)[-1].strip()

    text = f"User: {context}\nBhoomi: {reply}"

    output.append({"text": text})

print("Total clean samples:", len(output))
```

```python
with open("finetune_data.jsonl", "w", encoding="utf-8") as f:
    for item in output:
        f.write(json.dumps(item, ensure_ascii=False) + "\n")

print("✅ finetune_data.jsonl created")
```
